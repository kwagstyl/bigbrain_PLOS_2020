```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import subprocess
```

```python
resolutions = [20,40,100,200,300,400, 1000]
```

```python
profiledir = '/data1/users/kwagstyl/bigbrain/NeuralNetworks/TrainingData/TrainingData_lowres/'
```

```python
def load_profile(filename, index):
    """load specific profile from file"""
    with open(filename, 'r') as f:
        for k,line in enumerate(f):
            if k==index:
                profile=list(map(float,line.split(' ')))
                return np.array(profile)
```

```python
index =50
fig, axes = plt.subplots(1,1,figsize=(10,8))

for k,r in enumerate(resolutions):
    profilename=os.path.join(profiledir, 'training_'+str(r)+'_geo.txt')
    p=load_profile(profilename,index)
    if r != 20:
        p=65535-p
    axes.plot(p+12000*k+4000)
axes.set_xlim(0,200)
axes.set_yticks([])
axes.set_ylabel('Image intensity',size=20)
axes.set_xlabel('Cortical layers',size=20)
labelname=os.path.join(profiledir, 'training_cls.txt')
l=load_profile(labelname,index)
_,transitions=np.unique(l,return_index=True)
transitions=np.append(transitions,200)
colours=['#FFFFFF','#FFCCCC','#FFCC99','#FFFFCC','#CCFFCC','#CCFFFF','#99CCFF','#CCCCFF','#FFCCFF']
for k,l_i in enumerate(transitions):
    if k == len(transitions)-1:
        break
    axes.axvspan(transitions[k], transitions[k+1], alpha=0.5, color=colours[k])

axes.set_xticks([30,38,48,66,106,137,158,185])
axes.set_xticklabels(['pia','I', 'II','III','IV','V','VI','white matter'],size=15)

axes.legend(list(map(str,resolutions)),loc='upper left', title='Resolution $\mu$m$^3$',fontsize=10)
```
