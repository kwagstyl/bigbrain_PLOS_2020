```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm

import nibabel as nb
from scipy.stats import spearmanr, pearsonr
```

```python
#load in files
hemi='left'
Label=np.loadtxt('/data1/users/kwagstyl/bigbrain/NeuralNetworks/surfdir/medial_wall_'+hemi+'.txt')[:163842]
Curvatures=np.loadtxt('/data1/users/kwagstyl/bigbrain/NeuralNetworks/surfdir/mean_curvature_'+hemi+'_layer4.txt')[:163842]
Layers=np.array(['1', '2', '3', '4','5' ,'6'])
TotalThickness=np.loadtxt('/data1/users/kwagstyl/bigbrain/Hierarchy/thickness_'+hemi+'_total.txt')[:163842]
Thickness={}
for layer in range(6):
    Thickness["layer{0}".format(layer+1)]=np.loadtxt('/data1/users/kwagstyl/bigbrain/Hierarchy/thickness_'+hemi+'_layer'+str(layer+1)+'.txt')[:163842]
  
```

```python
#aggregate values by curvature for stacked bar chart
Sulc=-0.7
Gyr=0.4
Thresholds=np.array([[-100,Sulc],[Sulc, Gyr],[Gyr,100]])
#Thresholds=np.array([[-100,-1],[-1, -0.5],[-0.5,0.5],[0.5,1],[1,100]])
n_bins=20
Thresholds=np.linspace(-1.2,1.2,n_bins+1)

means=np.zeros((6,n_bins))
std=np.zeros((6,n_bins))
for n in range(6):
    i=-1
    for L in range(n_bins):
        i+=1
        Region=np.where(np.logical_and(Curvatures>Thresholds[L],Curvatures<Thresholds[L+1]))[0]
        means[n,i]=np.mean(Thickness["layer{0}".format(n+1)][Region])
        std[n,i]=np.std(Thickness["layer{0}".format(n+1)][Region])
```

```python
Layers=['I', 'II','III', 'IV', 'V', 'VI']
N=n_bins
ind=np.arange(N)/2.0
width=0.5
#colours=256-np.array([50,180,100,200,130,180,230,120,150])
colours=256-np.array([50,200,150,250,200,100])
cmap=plt.get_cmap('bone')
Plots={}
bottom=np.zeros(n_bins)
f=plt.figure()
for n in 5-np.arange(6):
    Plots["p{0}".format(n+1)] = plt.bar(ind, means[n,:], width, color=cmap(colours[n]),bottom=bottom)
    
    bottom+=means[n,:]
plt.xticks([N/16,N/4.5,3*N/8],['Fundus','Wall','Crown'])
plt.ylabel('Cortical thickness (mm)')
plt.tick_params(
    axis='x',          
    which='both',      
    bottom=False,      
    top=False)
plt.legend((Plots['p1'],Plots['p2'],Plots['p3'],Plots['p4'],Plots['p5'],Plots['p6']),Layers,title='Layer',bbox_to_anchor=(1,0.8))

f.savefig("LayersV1_"+hemi+".pdf", bbox_inches='tight')
```

```python
totals=np.sum(means,axis=0)
```

```python
Layers=['I', 'II','III', 'IV', 'V', 'VI']
N=n_bins
ind=np.arange(N)/2.0
width=0.5
colours=256-np.array([50,200,150,250,200,100])
cmap=plt.get_cmap('bone')
f=plt.figure()

Plots={}
bottom=np.zeros(N)
for n in 5-np.arange(6):
    Plots["p{0}".format(n+1)] = plt.bar(ind, means[n,:]/totals, width, color=cmap(colours[n]),bottom=bottom)
    
    bottom+=means[n,:]/totals
plt.xticks([N/16,N/4.5,3*N/8],['Fundus','Wall','Crown'])
plt.ylabel('Fractional cortical depth')

plt.tick_params(
    axis='x',          
    which='both',      
    bottom=False,      
    top=False)
plt.legend((Plots['p1'],Plots['p2'],Plots['p3'],Plots['p4'],Plots['p5'],Plots['p6']),Layers,title='Layer',bbox_to_anchor=(1,0.8))

f.savefig("LayersV1Normalised"+hemi+".pdf", bbox_inches='tight')
```
